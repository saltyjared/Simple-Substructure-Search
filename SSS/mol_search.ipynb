# Chem 274A Final Project: Molecular Substructure Searching in Python

Importing all required modules for the Molecule class and fingerprinting...

```python
from molecule import Molecule
```

Let's create a Molecule object from my favorite chemical compound, ***TNT***!

```python
tnt = Molecule("sdf/tnt.sdf") # Constructs a Molecule object from a .sdf file
tnt # Demonstrates that our _ipython_display_() method is working correctly, as a graphical representation of TNT appears below
```

From above, we see a few functional groups that may be of interest:
- Nitro groups ($NO_{2}$)
- Benzene ring
- Alkenes ($C=C$)

Let's look for them using both substructure searching methods (via SDF or via SMILES)!

```python
# Substructure searching via SDF
nitrogen_dioxide = Molecule("sdf/no2.sdf") # SDF representation of a nitro group
benzene = Molecule("sdf/benzene.sdf")
ethylene = Molecule("sdf/ethylene.sdf") # SDF representation of an alkene

# Using check_substructure_sdf method with our above substructures as inputs
print("TNT contains the nitro group as a substructure:", tnt.check_substructure_sdf(nitrogen_dioxide))
print("TNT contains benzene as a substructure:", tnt.check_substructure_sdf(benzene))
print("TNT contains ethylene as a substructure:", tnt.check_substructure_sdf(ethylene))
```

Great! We confirmed the presence of the above substructures via SDF representations. But what if we don't want to download the SDF files and want to quickly search for a substructure? We can use the SMILES representation of a bond path instead!

```python
# Substructure searching via SMILES and check_substructure_smiles()

print("TNT contains the nitro group as a substructure:", tnt.check_substructure_smiles("ON=O"))
print("TNT contains benzene as a substructure:", tnt.check_substructure_smiles("CC=CC=CC"))
print("TNT contains ethylene as a substructure:", tnt.check_substructure_smiles("C=C"))
```

Wonderful! Both SMILES and SDF files are compatible with our Molecule object and substructures can be found via either method. Just as a sanity check, let's search for a functional group that is NOT present in TNT to confirm the functionality of the above methods.

```python
# Using acetylene (C#C [triple bond]) as the substructure of interest

acetylene = Molecule("sdf/acetylene.sdf")
print("TNT contains acetylene as a substructure (via SDF):", tnt.check_substructure_sdf(acetylene))
print("TNT contains acetylene as a substructure (via SMILES):", tnt.check_substructure_smiles("C#C"))
```

Great! We are able to correctly search for substructures and confirm the absence of other substructures!
